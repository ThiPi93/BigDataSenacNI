```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
imdb_rating = np.genfromtxt('/content/imdb_projeto.csv', delimiter=';', skip_header=1, dtype=None, encoding='utf-8', usecols=8)
print(imdb_rating)
print(type(imdb_rating))

# Calcule a média:
media = np.mean(imdb_rating)
print(f"Média de notas no IMDB de candidatos ao Oscar de melhor filme: {media:.2f}")


# Obtenha a mediana:
mediana = np.median(imdb_rating)
print(f"Mediana de notas no IMDB de candidatos ao Oscar de melhor filme:{mediana:.2f}")


# Calcule a distância entre a média e a mediana:
distancia = (media - mediana) / mediana
print(f"Distância entre a média e a mediana: {distancia * 100:.2f}%")

if abs(distancia) <= 0.10:
    print("A média tende a ser uma medida de tendência central confiável.")
elif abs(distancia) < 0.25:
    print("A média pode estar sofrendo uma influência moderada de valores extremos.")
else:
    print("A média tende a não ser uma medida de tendência central confiável.")


if media > mediana:
    print("A influência é dos valores mais altos da distribuição.")
elif media < mediana:
    print("A influência é dos valores mais baixos da distribuição.")
```

```
import matplotlib.pyplot as plt  # Importando a biblioteca Matplotlib
import seaborn as sns  # Importando a biblioteca Seaborn

sns.histplot(imdb_rating, kde=True) # kde=True adiciona a curva de densidade, traduzindo seria "Kernel Density Estimate"
plt.title('Média de notas no IMDB de candidatos ao Oscar de melhor filme de 1927 à 2020')
plt.show()
```

```
q1 = np.percentile(imdb_rating, 25)
q2 = np.percentile(imdb_rating, 50)
q3 = np.percentile(imdb_rating, 75)

print(f"\nPrimeiro Quartil (Q1):{q1:.2f}")
print(f"Segundo Quartil (Mediana/Q2): {q2:.2f}")
print(f"Terceiro Quartil (Q3):{q3:.2f}")

if q1 <= media <= q3:
    print("\nA média está DENTRO da faixa interquartil (entre Q1 e Q3).")
else:
    print("\nA média está FORA da faixa interquartil (fora de Q1 e Q3).")

plt.axvline(x=q1, color='green', linestyle='--', label=f'Q1: R$ {q1:.2f}')
plt.axvline(x=q3, color='orange', linestyle='--', label=f'Q3: R$ {q3:.2f}')
plt.axvline(x=media, color='red', linestyle='-', label=f'Média: R$ {media:.2f}')
plt.axvline(x=mediana, color='purple', linestyle='-', label=f'Mediana: R$ {mediana:.2f}')
plt.legend()
plt.show()
```

```
from os import read
# Calcular Q1 e Q3
Q1 = np.percentile(imdb_rating, 25)
Q3 = np.percentile(imdb_rating, 75)

# Calcular IQR
IQR = Q3 - Q1

# Calcular Limites
limite_superior = Q3 + (1.5 * IQR)
limite_inferior = Q1 - (1.5 * IQR)

print(f"\n--- Limites de Outliers (IMDB Rating) ---")
print(f"Q1 (25%): R$ {Q1:.2f}")
print(f"Q3 (75%): R$ {Q3:.2f}")
print(f"IQR: R$ {IQR:.2f}")
print(f"Limite Superior (LS): R$ {limite_superior:.2f}")
print(f"Limite Inferior (LI): R$ {limite_inferior:.2f}")

df = pd.read_csv(
    '/content/imdb_projeto.csv',
    sep=';',
    engine='python',
    quoting=3
)

# Limpar nomes das colunas
df.columns = df.columns.str.replace('"', '').str.strip()

outliers_superiores = df[df['IMDB Rating'] > limite_superior]
outliers_inferiores = df[df['IMDB Rating'] < limite_inferior]

print(f"\n--- Outliers Superiores(IMDB Rating) ---")
print(outliers_superiores[['Film', 'IMDB Rating']].sort_values(
    by='IMDB Rating', ascending=False
))
print(f"\n--- Outliers Inferiores(IMDB Rating) ---")
print(outliers_inferiores[['Film', 'IMDB Rating']].sort_values(
    by='IMDB Rating', ascending=True
))

```

```
if not outliers_inferiores.empty or not outliers_superiores.empty:

    fig, axes = plt.subplots(1, 2, figsize=(14, 6)) # 1 linha, 2 colunas

    # 1ª Posição: Outliers Inferiores (Crescente)
    axes[0].bar(outliers_inferiores['Film'], outliers_inferiores['IMDB Rating'])
    axes[0].set_title('Outliers Inferiores (IMDB Rating)')
    axes[0].set_ylabel('Notas IMDB')
    axes[0].tick_params(axis='x', rotation=45, labelsize=8)
    axes[0].grid(axis='y', linestyle='--')

# 2ª Posição: Outliers Superiores (Decrescente)
    # Ordenamos novamente para garantir a visualização correta
    outliers_superiores_plot = outliers_superiores.sort_values(by='IMDB Rating', ascending=False)
    axes[1].bar(outliers_superiores_plot['Film'], outliers_superiores_plot['IMDB Rating'])
    axes[1].set_title('Outliers Superiores (IMDB Rating)')
    axes[1].set_ylabel('Notas IMDB')
    axes[1].tick_params(axis='x', rotation=45, labelsize=8)
    axes[1].grid(axis='y', linestyle='--')

    plt.tight_layout() # Ajusta automaticamente os parâmetros de subplot para dar preenchimento
    plt.show()

else:
    print("\nNão houve outliers superiores ou inferiores para plotar.")
```
