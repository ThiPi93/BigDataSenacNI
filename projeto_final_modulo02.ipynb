### Carregamento das Notas do IMDB

Nesta célula, os dados referentes às notas do IMDB são carregados diretamente
do arquivo CSV utilizando a função `genfromtxt()` do NumPy.
Foi selecionada especificamente a coluna referente às avaliações,
permitindo uma análise estatística focada apenas na variável numérica de interesse.
### Cálculo da Média Aritmética

A média aritmética das notas do IMDB é calculada utilizando a função `mean()`.
Essa medida representa o valor médio das avaliações atribuídas aos filmes
indicados ao Oscar de Melhor Filme, sendo uma importante medida de tendência central.
### Cálculo da Mediana

A mediana é calculada para identificar o valor central da distribuição das notas,
dividindo os dados em duas partes iguais.
Essa medida é especialmente útil quando há suspeita de influência de valores extremos,
pois não sofre impacto direto dos outliers.
### Distância entre Média e Mediana

Nesta etapa é calculada a distância percentual entre a média e a mediana.
Essa análise permite avaliar se a média é uma medida de tendência central confiável
ou se está sendo influenciada por valores extremos da distribuição.
### Análise da Influência na Distribuição

Com base na comparação entre média e mediana, é possível identificar
se a distribuição sofre maior influência de valores mais altos ou mais baixos,
indicando uma possível assimetria positiva ou negativa.


```
imdb_rating = np.genfromtxt('/content/imdb_projeto.csv', delimiter=';', skip_header=1, dtype=None, encoding='utf-8', usecols=8)
print(imdb_rating)
print(type(imdb_rating))

# Calcule a média:
media = np.mean(imdb_rating)
print(f"Média de notas no IMDB de candidatos ao Oscar de melhor filme: {media:.2f}")


# Obtenha a mediana:
mediana = np.median(imdb_rating)
print(f"Mediana de notas no IMDB de candidatos ao Oscar de melhor filme:{mediana:.2f}")


# Calcule a distância entre a média e a mediana:
distancia = (media - mediana) / mediana
print(f"Distância entre a média e a mediana: {distancia * 100:.2f}%")

if abs(distancia) <= 0.10:
    print("A média tende a ser uma medida de tendência central confiável.")
elif abs(distancia) < 0.25:
    print("A média pode estar sofrendo uma influência moderada de valores extremos.")
else:
    print("A média tende a não ser uma medida de tendência central confiável.")


if media > mediana:
    print("A influência é dos valores mais altos da distribuição.")
elif media < mediana:
    print("A influência é dos valores mais baixos da distribuição.")
```

### Cálculo dos Quartis

Os quartis Q1, Q2 (mediana) e Q3 são calculados para dividir a distribuição
em quatro partes iguais. Essa etapa é fundamental para compreender
a dispersão dos dados e preparar a análise de outliers.
### Representação Gráfica dos Quartis, Média e Mediana

São adicionadas linhas verticais ao gráfico para representar
Q1, Q3, média e mediana, facilitando a comparação visual
entre medidas de tendência central e dispersão.


```
q1 = np.percentile(imdb_rating, 25)
q2 = np.percentile(imdb_rating, 50)
q3 = np.percentile(imdb_rating, 75)

print(f"\nPrimeiro Quartil (Q1):{q1:.2f}")
print(f"Segundo Quartil (Mediana/Q2): {q2:.2f}")
print(f"Terceiro Quartil (Q3):{q3:.2f}")

if q1 <= media <= q3:
    print("\nA média está DENTRO da faixa interquartil (entre Q1 e Q3).")
else:
    print("\nA média está FORA da faixa interquartil (fora de Q1 e Q3).")

plt.axvline(x=q1, color='green', linestyle='--', label=f'Q1: R$ {q1:.2f}')
plt.axvline(x=q3, color='orange', linestyle='--', label=f'Q3: R$ {q3:.2f}')
plt.axvline(x=media, color='red', linestyle='-', label=f'Média: R$ {media:.2f}')
plt.axvline(x=mediana, color='purple', linestyle='-', label=f'Mediana: R$ {mediana:.2f}')
plt.legend()
plt.show()
```

### Cálculo do Intervalo Interquartil (IQR)

O IQR é calculado pela diferença entre o terceiro e o primeiro quartil.
Essa medida é amplamente utilizada para identificar valores extremos
e compreender a dispersão central da distribuição.
### Definição dos Limites para Outliers

Com base no IQR, são definidos os limites superior e inferior
utilizando o critério de 1,5 vezes o intervalo interquartil.
Esses limites permitem identificar observações consideradas atípicas.



```
from os import read
# Calcular Q1 e Q3
Q1 = np.percentile(imdb_rating, 25)
Q3 = np.percentile(imdb_rating, 75)

# Calcular IQR
IQR = Q3 - Q1

# Calcular Limites
limite_superior = Q3 + (1.5 * IQR)
limite_inferior = Q1 - (1.5 * IQR)

print(f"\n--- Limites de Outliers (IMDB Rating) ---")
print(f"Q1 (25%): R$ {Q1:.2f}")
print(f"Q3 (75%): R$ {Q3:.2f}")
print(f"IQR: R$ {IQR:.2f}")
print(f"Limite Superior (LS): R$ {limite_superior:.2f}")
print(f"Limite Inferior (LI): R$ {limite_inferior:.2f}")

df = pd.read_csv(
    '/content/imdb_projeto.csv',
    sep=';',
    engine='python',
    quoting=3
)

# Limpar nomes das colunas
df.columns = df.columns.str.replace('"', '').str.strip()

outliers_superiores = df[df['IMDB Rating'] > limite_superior]
outliers_inferiores = df[df['IMDB Rating'] < limite_inferior]

print(f"\n--- Outliers Superiores(IMDB Rating) ---")
print(outliers_superiores[['Film', 'IMDB Rating']].sort_values(
    by='IMDB Rating', ascending=False
))
print(f"\n--- Outliers Inferiores(IMDB Rating) ---")
print(outliers_inferiores[['Film', 'IMDB Rating']].sort_values(
    by='IMDB Rating', ascending=True
))

```

### Visualização Gráfica dos Outliers

Nesta célula são gerados gráficos de barras para visualizar
os outliers superiores e inferiores, facilitando a interpretação
dos valores extremos encontrados na análise.


```
if not outliers_inferiores.empty or not outliers_superiores.empty:

    fig, axes = plt.subplots(1, 2, figsize=(14, 6)) # 1 linha, 2 colunas

    # 1ª Posição: Outliers Inferiores (Crescente)
    axes[0].bar(outliers_inferiores['Film'], outliers_inferiores['IMDB Rating'])
    axes[0].set_title('Outliers Inferiores (IMDB Rating)')
    axes[0].set_ylabel('Notas IMDB')
    axes[0].tick_params(axis='x', rotation=45, labelsize=8)
    axes[0].grid(axis='y', linestyle='--')

# 2ª Posição: Outliers Superiores (Decrescente)
    # Ordenamos novamente para garantir a visualização correta
    outliers_superiores_plot = outliers_superiores.sort_values(by='IMDB Rating', ascending=False)
    axes[1].bar(outliers_superiores_plot['Film'], outliers_superiores_plot['IMDB Rating'])
    axes[1].set_title('Outliers Superiores (IMDB Rating)')
    axes[1].set_ylabel('Notas IMDB')
    axes[1].tick_params(axis='x', rotation=45, labelsize=8)
    axes[1].grid(axis='y', linestyle='--')

    plt.tight_layout() # Ajusta automaticamente os parâmetros de subplot para dar preenchimento
    plt.show()

else:
    print("\nNão houve outliers superiores ou inferiores para plotar.")
```

### Medidas de Dispersão

Nesta etapa são calculadas variância, desvio padrão
e coeficiente de variação, fornecendo uma análise
quantitativa do grau de dispersão dos dados.


```
imdb_rating = df['IMDB Rating']
media_imdb_rating = imdb_rating.mean()
variancia_imdb_rating = np.var(imdb_rating, ddof=0)
desvio_padrao_rating = np.std(imdb_rating, ddof=0)

cv_rating = (desvio_padrao_rating/media_imdb_rating)*100
distancia_rating = variancia_imdb_rating / (media_imdb_rating ** 2)

print(f"--- Análise dos Ratings Totais candidatos ao Oscar de Melhor Filme (1927 à 2020) ---")
print(f"\nMédia dos Filmes Totais: {media_imdb_rating:.2f}")
print(f"Variância dos Filmes Totais: {variancia_imdb_rating:.2f}")
print(f"Desvio Padrão dos Ratings Totais: {desvio_padrao_rating:.2f}")
print("-" * 100)
print(f"Coeficiente de Variação (CV): {cv_rating:.2f}%")
print(f"Distância da Variância / Média²: {distancia_rating:.2f}")

# Análise de Dispersão
if distancia_rating <= 0.10:
    analise_rating = "Baixa dispersão dos dados em relação à média."
elif distancia_rating < 0.25:
    analise_rating = "Dispersão moderada dos dados em relação à média."
else:
    analise_rating = "Alta dispersão dos dados em relação à média."

print(f"Conclusão da Dispersão: {analise_rating}")
```

### Análise de Assimetria

A assimetria da distribuição é analisada para identificar
se os dados são simétricos, assimétricos positivos
ou negativos, auxiliando na compreensão da forma dos dados.
### Análise de Curtose

A curtose é calculada para avaliar o grau de concentração
dos dados em torno da média e o comportamento das caudas
da distribuição, indicando se ela é mesocúrtica,
platicúrtica ou leptocúrtica.


```
## Dataframe
imdb_df = pd.read_csv(
    '/content/imdb_projeto.csv',
    sep=';',
    engine='python',
    quoting=3,
    encoding='latin1'
)

imdb_df.columns = imdb_df.columns.str.replace('"', '').str.strip()

rating_imdb = imdb_df['IMDB Rating']

## Calcular Assimetria
assimetria = rating_imdb.skew()
print(f"Assimetria das notas no IMDB: {assimetria:.4f}")

media = rating_imdb.mean()
mediana = rating_imdb.median()
print(f"Média: {media:.2f}")
print(f"Mediana: {mediana:.2f}")

if assimetria >= -0.5 and assimetria <= 0.5:
    analise_assimetria = "Simétrica (ou Quase Simétrica). Média e Mediana são próximas."
elif assimetria > 0.5:
    analise_assimetria = "Positiva. A cauda se estende para a direita (valores maiores). Média > Mediana."
else:
    analise_assimetria = "Negativa. A cauda se estende para a esquerda (valores menores). Média < Mediana."

print(f"\nConclusão da Assimetria: {analise_assimetria}")

## Calcular Curtose
curtose_excesso = rating_imdb.kurtosis()
curtose_real = curtose_excesso + 3
print(f"Curtose em Excesso (Pandas): {curtose_excesso:.4f}")
print(f"Curtose Real (Referência 3.0): {curtose_real:.4f}")

if curtose_real >= 2.5 and curtose_real <= 3.5:
    analise_curtose = "Mesocúrtica. Distribuição próxima da normal (dados uniformes no entorno da média)."
elif curtose_real < 2.5:
    analise_curtose = "Platicúrtica. Dados mais dispersos em relação à média. Caudas finas e Outliers comuns."
else: # curtose_real > 3.5
    analise_curtose = "Leptocúrtica. Dados extremamente concentrados no centro e caudas pesadas. Outliers muito comuns."

print(f"\nConclusão da Curtose: {analise_curtose}")
```
